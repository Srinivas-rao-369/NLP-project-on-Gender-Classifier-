```python
import re
from nltk import word_tokenize
```

```python
def cleaned_text(text):
    text = str(text)
    text = text.lower()
    cleaned_text = re.sub('[^a-zA-Z0-9]', ' ', text)
    list_of_words = word_tokenize(cleaned_text)
    clean_text = ''
    clean_text = [clean_text + ' ' + each_word.lower() for each_word in list_of_words]
    
    clean_text = re.sub('\s\W',' ',text)
    clean_text = re.sub('\W\s',' ',text)
        
    clean_text = re.sub('\s+',' ',text)
    clean_text = re.sub('[^\x00-\x7F]+',' ', text)
    clean_text = re.sub('https?:\/\/.*[\r\n]*', ' ', text)
    clean_text = re.sub('[?!+%{}:;.,"\'()\[\]_]', '',text)
    clean_text = re.sub('\s+',' ',text)
    return clean_text
```

```python
data['clean_text'] = [cleaned_text(each_text) for each_text in data['text']]
```

```python
data['clean_description'] = [cleaned_text(each_text) for each_text in data['description']]
```

```python
independent_cols = data[['_golden', '_unit_state', '_trusted_judgments',
       'gender:confidence', 'profile_yn', 'profile_yn:confidence', 'created',
       'fav_number', 'link_color', 'name', 'profileimage', 'retweet_count',
       'sidebar_color', 'tweet_count', 'tweet_id', 'clean_text',
       'clean_description']].values

dependent_cols = data['gender'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent_cols, dependent_cols, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()
```

```python
y_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```
